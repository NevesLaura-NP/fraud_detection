Let's take a further look between the difference between flagged frauds and actual frauds by ploting a barplot maybe then we can recognize the current efficiency of the fraud flagging

```python
fraud_counts = df["isFraud"].value_counts()
flagged_fraud_counts = df["isFlaggedFraud"].value_counts()


fig, ax = plt.subplots()
fraud_counts.plot(kind='bar', ax=ax, position=0, width=0.4, label='Fraud')
flagged_fraud_counts.plot(kind='bar', ax=ax, position=1, width=0.4, color='purple', label='Flagged Fraud')


ax.set_xlabel('Frauds')
ax.set_ylabel('Count')
ax.legend( 'Frauds vs Flagged Frauds')

plt.show()
```

After taking a look at the barplot we can regard that a good majority of the transactions are not fraud and that the small ammount of frauds is bigger than the Flagged Frauds, since the ammount between Frauds and not not Frauds is great we need to take a further look into the discrepancy between Frauds and successfly Flagged Frauds

```python
fraud_counts = df["isFraud"].eq(1).sum()
flagged_fraud_counts = df["isFlaggedFraud"].eq(1).sum()


fig, ax = plt.subplots()
ax.bar([0, 1], [fraud_counts, flagged_fraud_counts], width=0.4, color=['blue', 'purple'])


ax.set_xticks([0, 1])
ax.set_xticklabels(['Fraud', 'Flagged Fraud'])
ax.set_xlabel('Frauds')
ax.set_ylabel('Count')
ax.set_title('Count of Frauds and Flagged Frauds')

plt.show()
```

After plotting the count of Frauds vs succesfully flagged Frauds we see there is a difference of thousands of transactions. Let's take a look at some other variables in other to understand what is missing from the flagging system. Right now we can drop the Flagged Fraud column to agilize our analysis and stablish only Frauds are our idependent analysis.

```python
df = df.drop("isFlaggedFraud", axis = 1)
```

Let's take a further look into the other variables and discover if there is any potential pattern with existant frauds. I will be starting with types of transfer. I will be using a lollipop plot to visualize also the counting of the total of transactions of each transfer type that are fraudulant.

```python
fraudulent_transactions = df[df['isFraud'] == 1]
fraud_account_type_counts = fraudulent_transactions.groupby('type').size().reset_index(name='counts')


sort_data = fraud_account_type_counts.sort_values(by='counts', ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))
ax.vlines(x=sort_data.index, ymin=0, ymax=sort_data['counts'], color='red', alpha=0.7, linewidth=2)
ax.scatter(x=sort_data.index, y=sort_data['counts'], s=75, color='red', alpha=0.7)

ax.set_title('Plot of Fraudulent Payments Vs Types')
ax.set_xlabel('Account Type')
ax.set_ylabel('Count')

for index, value in zip(sort_data.index, sort_data['counts']):
    ax.text(index, value + 0.1, f'{value}', ha='center', va='bottom', fontsize=10, color='black')

ax.set_xticks(sort_data.index)
ax.set_xticklabels(sort_data['type'], rotation=45, ha='right')

plt.show()
```

We already can see from this plot that frauds are committed only through transfers and cash out types. Lets take a look at these compared to the total ammount of transfer and cashout occurances 

```python

fraud_account_type_counts = df.groupby('type').size().reset_index(name='counts')


sort_data = fraud_account_type_counts.sort_values(by='counts', ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))
ax.vlines(x=sort_data.index, ymin=0, ymax=sort_data['counts'], color='blue', alpha=0.7, linewidth=2)
ax.scatter(x=sort_data.index, y=sort_data['counts'], s=75, color='blue', alpha=0.7)

ax.set_title('Plot of Total Payments from Each Account Types')
ax.set_xlabel('Account Type')
ax.set_ylabel('Count')

for index, value in zip(sort_data.index, sort_data['counts']):
    ax.text(index, value + 0.1, f'{value}', ha='center', va='bottom', fontsize=10, color='black')

ax.set_xticks(sort_data.index)
ax.set_xticklabels(sort_data['type'], rotation=45, ha='right')

plt.show()
```

We can observe now how even within the Cash Out and Transfer types there is are at least hundreds of thousands of payments that are not fraud, however we can also regard there are quite unequal distribution of the payment types. From now on to visualize other aspects of the dataframe we need start utilizing a sample size as the whole data is too big and therefore computetionally expensive. For the following correlation matrix we will take a sample of 3000 rows and incorporate more exploratory analysis to it. I chose 3000 as the sample size needs to be somewhat proportionate to the size of the entire data.

```python
sample_df = df.sample(n=3000, replace=False)

sample_df=sample_df.drop("type", axis=1)

# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(sample_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
log_amount = np.log(sample_df['amount'])

plt.figure(figsize=(10, 6))
sns.histplot(log_amount, bins=50, kde=True)
plt.title('Distribution of Payment Amounts')
plt.show()
```

```python
plt.figure(figsize=(19, 6))
sns.boxplot(x='isFraud', y=log_amount, data=sample_df)
plt.title('Amount Distribution for Fraudulent Transactions')
plt.show()
```

```python
fraudulent_transaction = df[df["isFraud"] == 1]

plt.figure(figsize=(12, 6))
sns.scatterplot(x='oldbalanceOrg', y='newbalanceOrig', data=fraudulent_transaction, alpha=0.5)
plt.title('Old Balance vs. New Balance for Fraudulent Transactions')
plt.xlabel('Old Balance')
plt.ylabel('New Balance')
plt.show()
```
